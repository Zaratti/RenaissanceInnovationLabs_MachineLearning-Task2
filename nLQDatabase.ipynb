```python
import json
from sqlalchemy import create_engine, text
from sqlalchemy.exc import SQLAlchemyError
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import io
import base64

# Step 1: Create the engine (make sure to replace with actual connection details)
DATABASE_FILE = "example_database.sqlite"
engine = create_engine(f"sqlite:///{DATABASE_FILE}")

# Step 2: We used checkTableStructure to Whitelist columns to prevent SQL injection
VALID_COLUMNS = {'id', 'name', 'email', 'created_at'}

def validate_query_columns(columns):
    # Ensure all columns in the query are valid
    return all(column in VALID_COLUMNS for column in columns)

# Step 3: Define the function to execute queries
def execute_sql_query(sql_query: str, params: dict = None):
    try:
        with engine.connect() as connection:
            # Use `text()` to safely execute raw SQL with parameters
            result = connection.execution_options(stream_results=True).execute(text(sql_query), params or {})
            # Fetch all results and format as dictionaries (e.g., list of dicts)
            return [dict(row._mapping) for row in result]
            # Using SQLAlchemyError in place of Exception to catch errors
    except SQLAlchemyError as e:
        print("Database error:", str(e))
        return None

# Step 4: Data Formating and Response handling
def process_results(results, format_type="text"):
    if format_type == "text":
        return json.dumps(results, indent=2)
    
    elif format_type == "table":
        return json.dumps(results)
        
    elif format_type == "chart":
        # Example: plotting user count over time based on created_at dates
        dates = [row["created_at"] for row in results]
        plt.figure(figsize=(10, 6))
        plt.hist(dates, bins=10)
        plt.xlabel("Date")
        plt.ylabel("User Count")
        plt.title("User Signups Over Time")
        
        # Save chart to a bytes buffer and encode in base64
        buf = io.BytesIO()
        plt.savefig(buf, format="png")
        buf.seek(0)
        img_base64 = base64.b64encode(buf.read()).decode("utf-8")
        buf.close()
        plt.close()
        
        # Return as an embedded image
        return f"data:image/png;base64,{img_base64}"
    else:
        return "Invalid format type"

# Define query based on sanitized columns
columns = ['id', 'name', 'created_at']  # Example selected columns
if validate_query_columns(columns):
    query = f"SELECT {', '.join(columns)} FROM users WHERE created_at > :min_date"
    results = execute_sql_query(query, {"min_date": "2021-01-01"})

    # Choose a format: "text", "table", or "chart"
    formatted_response = process_results(results, format_type="text")
    print(formatted_response)  # Output the result based on the chosen format
else:
    print("Invalid columns requested")
```

```python
# Used this function to check table structure
def check_table_structure():
    with engine.connect() as connection:
        result = connection.execute(text("PRAGMA table_info(users);"))
        # Use row[1] to access the column name in each tuple
        columns = [row[1] for row in result]
        print("Columns in 'users' table:", columns)

check_table_structure()
```

```python
# Step 3: Call the function with a sample query to filter users based on `created_at`
query = "SELECT * FROM users WHERE created_at > :min_date"
# Calculate a date 5 years ago from today for example
five_years_ago = datetime.now() - timedelta(days=5*365)
results = execute_sql_query(query, {"min_date": five_years_ago.strftime('%Y-%m-%d')})

# Check the results
if results is not None:
    print(results)
else:
    print("No matching records found.")
```
